```python
# Предобработка текстов
def tokenize_lemma_stopwords(text):
    morph = MorphAnalyzer()
    stopword = set(stopwords.words('russian'))
    tokens = nltk.tokenize.word_tokenize(text.lower())
    tokens = [t for t in tokens if t.isalpha()]
    tokens = [morph.parse(t)[0].normal_form for t in tokens]
    tokens = [t for t in tokens if len(t) > 2]
    tokens = [t for t in tokens if t not in stopword]
    return tokens

texts = [text1, text2, text3, text4, text5]
texts = [tokenize_lemma_stopwords(text) for text in texts]

dictionary = gensim.corpora.Dictionary(texts)
dictionary.filter_extremes(no_below=1)
bow_corpus = [dictionary.doc2bow(doc) for doc in texts]
```

```python
def LSA_model(num_of_topics):
    model = gensim.models.LsiModel(bow_corpus, id2word=dictionary, num_topics=num_of_topics)
    print(f'Результат работы LSA (num_of_topics={num_of_topics})', )
    pprint(model.print_topics())
```

```python
# попробуйте значения от 1 до 10
for num_of_topics in range(1, 10):
    LSA_model(num_of_topics) 
```

```python
def PLSA_model(num_of_topics):
    model = gensim.models.LdaModel\
    (
        bow_corpus,
        id2word=dictionary,
        num_topics=num_of_topics,
        offset=2,
        update_every=1,
        passes=10,
        alpha=1,
        eta=1,
        per_word_topics=True
    )
    print(f'Результат работы PLSA (num_of_topics={num_of_topics})')
    pprint(model.print_topics())
```

```python
# попробуйте значения от 1 до 10
for num_of_topics in range(1, 10):
    PLSA_model(num_of_topics) 
```

```python
def LDA_model(num_of_topics, alpha, eta):
    model = gensim.models.LdaModel\
    (
        bow_corpus,
        id2word=dictionary,
        num_topics=num_of_topics,
        offset=2,
        update_every=1,
        passes=10,
        alpha=alpha,   # можно попробовать alpha = 'auto'
        eta=eta,       # можно попробовать eta = 'auto'
        per_word_topics=True
    )
    print(f'Результат работы LDA (num_of_topics={num_of_topics}, alpha={alpha}, eta={eta})')
    pprint(model.print_topics())
```

```python
eta_100 = [100]*len(dictionary) # количество компонент вектора должно совпадать с количеством слов в словаре

#  попробуйте значения от 1 до 10
for num_of_topics in range(1, 10):
    alpha = [100]*num_of_topics # количество компонент вектора должно совпадать с количеством тем
    eta = eta_100
    LDA_model(num_of_topics, alpha, eta) 
    alpha = 'auto'
    LDA_model(num_of_topics, alpha, eta)
    eta = 'auto'
    LDA_model(num_of_topics, alpha, eta)
```
